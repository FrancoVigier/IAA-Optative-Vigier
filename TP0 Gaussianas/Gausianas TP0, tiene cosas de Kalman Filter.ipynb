### Ejercicio A:
Con la guia del apunte de Gaussianas, creo las medias/centros teorico de nuestras clases y creamos la matrix de cov diagonal que a efectos de la gráfica cada clase va a tener un heat map al crear sus puntos de una elipse con ejes paralelos a lo coordenados.

Sin mucho más como resultado a contraste del apartado B podemos ver la la varianza es mayor dado que la dev std es mayor y las medias del conjunto de puntos de cada clase se va a dispersar un poco de las teóricas planteadas en el enunciado. Si bien en este apartado el plotteo de ambas clases parece una elipse con ejes NO paralelo a los coordenados, si bajamos la dev std se pueden ver bien ambas clases(C= 0.05 y N= 20000) y que independientemente cada una es una elipse con ejes paralelos a los coordenados. Pero si 
quisieramos tener una elips con ejes no paralemos nuestra matrix de cov no debe ser diagonal pero en si por el último teorema del apunte de Gaussianas se podría encontrar una transformación natural entre las matrices

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

#Punto A del TP, creo las clases 1 y 2, se detalla más abajo
def gaussianClassesCovMatrixDiag (d,n, C):

#Creo los centros de los cuales las clases van a "girar", d-dimensionales el (1,....,1) y el (-1,....,-1)
    clase1Center = np.ones((d,), dtype=int)
    clase2Center = np.negative(clase1Center)
#Segun el apunte de gaussianas el caso especial de las matrices diagonales y en cada elemento (i,i) de la matriz
#está la varianza( dev estandar cuadrado) en cada variable, en este caso es sobre cada coordenada. Recordemos que al ser
# una matrix de cov Diagonal, se va a tener un heatmap de una elipse centrada en los ejes en cada Clase o paralelos a ambos ejes.
#Si disminuimos C lo suficiente se puede apreciar(0.1)
    devEst = (C*d*d)
    cov = np.diag(np.full(d,devEst*devEst))
#Genero con distribucion Gaussiana n/2 puntos, con la matrix de Covarianza Diagonal, con centro en sus respectivos puntos
    puntosClase1 = np.random.multivariate_normal(clase1Center,cov,n//2)
    puntosClase2 = np.random.multivariate_normal(clase2Center,cov,n//2)
#Creo los DataFrame de ambas Clases y los devuelvo. Lo pude haber agrupado pero ne
    columnas = [("Coor"+ str(i+1)) for i in range(d)] #Hago una label para las columas sino queda feo
    dataPandaC1 = pd.DataFrame(puntosClase1, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
    dataPandaC2 = pd.DataFrame(puntosClase2, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
#Los retorno
    return dataPandaC1, dataPandaC2
```

```python
#Testeo de la distribucion Gaussiana con Matrix de Cov Diagonal
clase1DF, clase2DF = gaussianClassesCovMatrixDiag(2,200,0.75) # con C = 0, hace sentido que el unico punto sea el Centro Mean
#Grafico ambas clases
ax1 = clase1DF.plot(kind='scatter', x='Coor1', y='Coor2', color='r', label = 'Clase1')  
ax2 = clase2DF.plot(kind='scatter', x='Coor1', y='Coor2', color='g', ax = ax1, label='Clase2')  
#Calculo la desv std, para las clases 1 y 2. Podemos ver que la desviación de los datos es cercano al propuesto en el enunciado
#en cadad columna de ambas clases. Como la desv est propuesta en el apartado A es a mi parecer un poco elevada, podemos aticiparnos
#que la media de los datos va a desplazarse un poco... con respecto a esa media teórica para ambas clases el (1,...,1) y el 
#(-1,...,-1) ya que la dev std marca 'que tan agrupados de la media teorica estan los datos'
stdC1 = clase1DF.std()
print("La desviación estandar para la Clase1 en sus columnas es:")
print(stdC1)

stdC2 = clase2DF.std()
print("La desviación estandar para la Clase2 en sus columnas es:")
print(stdC2)

#Acompañado de lo anterior vemos las medias de los datos y qué tan alejadas de la media teorica para cada clase están. Recordemos
#que la dev std es 3.
meanC1 = clase1DF.mean(axis = 0)
print("La media para la Clase1 en sus columnas es:")
print(meanC1)

meanC2 = clase2DF.mean(axis = 0)
print("La media para la Clase2 en sus columnas es:")
print(meanC2)

```

### Ejercicio B:

Básicamente como el anterior apartado y guiándome con el apunte de Gaussianass, voy a crear los centros de las clases esos centros son d-dimensionales y nuestros puntos van a "girar" entorno a ellos... Segun el ejercicio tenemos como dev std un C constante que armando la matix de cov diagonal pongo la varianza en sus elementos (i,i), como ahora la dev std es más chica que en el anterior apartado entonces las clases en su MEDIA va a ser cercana a sus respectivos centros. Luego creamos los dataframe correspondientes.

Acotando un poco podemos decir que ambas clases al tener una matrix cov diagonal se va a "generar" una elipse con ejes paralelos a los de coordenadas. Y si sumamos estas cualidades individuales a que los centros están sobre la misma recta paralela al eje x, las elipses que se formen se van a mezclar en una con las mismas cualidades ya mencionadas. Con un C relativamente chico un 0,20 o 0,10 podemos observar estas 2 elipses por separado. 

En la celda siguiente testeamos lo desarrollado, constatando las medias de las clases y la dev std de las mismas, viendo como se acercan a las mencionadas en el enunciado. Obviamente estas medidas de checkeo van a depender de la dev std perse y de la cantidad de puntos a samplear, pero si reducimo la std podemos ver que los puntos obviamente se acercan más a la media teorica propuesta para cada clase...

```python
#Punto B del TP, creo las clases 1 y 2, se detalla más abajo

def gaussianClassesCovMatrixDiagC (d,n, C):

#Creo los centros de los cuales las clases van a "girar", d-dimensionales el (1,....,0) y el (-1,....,0)
    clase1Center = np.zeros((d,), dtype=int)
    clase1Center[0] = 1
    clase2Center = np.negative(clase1Center)
#Segun el apunte de gaussianas el caso especial de las matrices diagonales y en cada elemento (i,i) de la matriz
#está la varianza( dev estandar cuadrado) en cada variable, en este caso es sobre cada coordenada. Recordemos que al ser
# una matrix de cov Diagonal, se va a tener un heatmap de una elipse centrada en los ejes en cada Clase
    devEst = C
    cov = np.diag(np.full(d,devEst*devEst))
#Genero con distribucion Gaussiana n/2 puntos, con la matrix de Covarianza Diagonal, con centro en sus respectivos puntos
    puntosClase1 = np.random.multivariate_normal(clase1Center,cov,n//2)
    puntosClase2 = np.random.multivariate_normal(clase2Center,cov,n//2)
#Creo los DataFrame de ambas Clases y los devuelvo. Lo pude haber agrupado pero ne
    columnas = [("Coor"+ str(i+1)) for i in range(d)] #Hago una label para las columas sino queda feo
    dataPandaC1 = pd.DataFrame(puntosClase1, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
    dataPandaC2 = pd.DataFrame(puntosClase2, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
#Los retorno
    return dataPandaC1, dataPandaC2
```

```python
#Testeo de la distribucion Gaussiana con Matrix de Cov Diagonal Con C como unico factor de la dev std
clase1DFonlyC, clase2DFonlyC = gaussianClassesCovMatrixDiagC (4,2000, 2.0)

#De forma hómologa al apartado A...
ax1onlyC = clase1DFonlyC.plot(kind='scatter', x='Coor1', y='Coor2', color='y', label = 'Clase1')  
ax2onlyC = clase2DFonlyC.plot(kind='scatter', x='Coor1', y='Coor2', color='b', ax = ax1onlyC, label='Clase2')  

stdC1onlyC = clase1DFonlyC.std()
print("La desviación estandar para la Clase1 en sus columnas es:")
print(stdC1onlyC)

stdC2onlyC = clase2DFonlyC.std()
print("La desviación estandar para la Clase2 en sus columnas es:")
print(stdC2onlyC)

#Acompañado de lo anterior vemos las medias de los datos y qué tan alejadas de la media teorica para cada clase están. Recordemos
#que la dev std es 2.0.
meanC1onlyC = clase1DFonlyC.mean(axis = 0)
print("La media para la Clase1 en sus columnas es:")
print(meanC1onlyC)

meanC2onlyC = clase2DFonlyC.mean(axis = 0)
print("La media para la Clase2 en sus columnas es:")
print(meanC2onlyC)

#Podemos ver que al ser todavia una matriz diagonal de covarianza, las clases van a ser elipses con sus ejes paralelos al x e y.
#como achicamos la dev std y por lo tanto la varianza podemos ver que la media de los puntos generados esta relativamente muy cerca
# de las medias teóricas propuettas para este apartado
```
