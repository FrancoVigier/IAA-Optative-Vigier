Como era de esperarse el arbol entrenado con 3000 puntos tiene un resultado más optimo que los demás, esto se debe a que al tener mayor cantidad de puntos para entrenar el arbol puede generar más nodos y niveles de labels. Como estamos en el plano cartesiano sabemos que las restricciones de los nodos son comparaciones aritmeticas que usan X y Y de los puntos. Por lo tanto esta fragmentacion del plano que va a hacer el arbol es ortogonal, siempre va a ser basada a rectas paralelas a los ejes coordenados. Por eso a menor puntos para entrenar, estas rectas y la ortogonalidad de la separacion de los planos es más evidente... sumesé los comentarios del codigo perse

##### TP0: Spiral

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.metrics import accuracy_score

import numpy as np
import pandas as pd

def SnailNestedSpiral(n):
    clase1 = []
    clase2 = []
    clasePuntos = []
    puntosC1 = 0
    puntosC2 = 0
    
    while puntosC1 < n//2 or puntosC2 < n//2 :
        ro = np.sqrt(np.random.uniform(low=0.0, high=1.0))
        phiCirc = np.random.uniform(low= 0.0, high= np.pi * 2)      
        x = ro * np.cos(phiCirc)
        y = ro * np.sin(phiCirc)
        if ((phiCirc) / (4*np.pi) < ro and ro < ((phiCirc)+np.pi) / (4*np.pi)) or ((phiCirc + 2*np.pi) / (4*np.pi) < ro and ro < ((phiCirc + 2*np.pi)+np.pi) / (4*np.pi)) or ((phiCirc - 2*np.pi) / (4*np.pi) < ro and ro < ((phiCirc- 2*np.pi)+np.pi) / (4*np.pi)):
            if(puntosC1 < n//2):
                #clase1.append([x,y, "Clase1"])
                clasePuntos.append([x,y, "Clase1"])
                puntosC1 += 1
        else:
            if( puntosC2 < n//2):
                #clase2.append([x,y, "Clase2"])
                clasePuntos.append([x,y, "Clase2"])
                puntosC2 += 1

    columnas = [("Coor"+ str(i+1)) for i in range(2)] #Hago una label para las columas sino queda feo
    columnas.append("Clase")
    #dataClase1 = pd.DataFrame(clase1, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
    #dataClase2 = pd.DataFrame(clase2, columns = columnas) #Hago el dataFrame con los puntos de la clase2 y con la label
    dataPuntos =  pd.DataFrame(clasePuntos, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
    return dataPuntos
```

### EJERCICIO 2: Overfitting
   Explicando un poco del ejercicio cuando entrenamos con el train set de las elipses diagonales, a priori podemos pensar sobre
la separacion de las mismas... a más separadas más fácil para un arbol segmentar el plano entre menos secciones y así poder tener una frontera clara entre ambas clases y una cantidad relativamente baja de nodos. Pero en este caso tenemos solapamiento entre las elipses diagonales entonces el arbol va a intentar partir el plano con una "diagonal" por eso podemos ver como ese fenómeno de una "escalera" que se forma e intenta separar ambas clases obviamente a más grande el set train mejor se va a poder
apreciar este fenómeno. Pero sabemos que esta separación es complicada ponerla en nodos que tengan solo condiciones de x e y, el camino para trazar "escalera" por así decirlo es mucha más costosa a nivel arbol y por eso podemos observar que lo nodos que usan los arbol de Diagonal son desde un inicio significativamente más pesados y más cantidad. Luego con mayor tamaño de N, error train set disminuye pero aún así el error en el Test Set a medida que incrementamos el N se "clava" alrededor del 0.1 y esto no es tan bueno porque a incrementar N como TestError se clavó pero TrainError sigue reduciendose podemo decir que el modelo está aprendiendo el conjunto de Train, se está sobreentrenando.
Para los mismos valores de generacion de puntos pasa lo opuesto en las elipses PARALELAS, ya que intuitivamente podemos pensar de un inicio que podemos dividir ambas clases con la mediatriz entre sus centros, obviamente si no hay gran dispersion entre sus puntos. Esta es la idea supongo que sigue el 2do conjunto de arboles, intentan dividir las clases por una recta paralela al eje Y que esté relativamente en el medio de los centros de las elipses. De esta forma relativamente sencilla podemos tener bajo error y la segmentación/division del plano cartesiano es más simple, por eso se refleja en un bajo error y en una estructura de arbol mucho más simple. Y para un mismo N el modelo tiene mejores resultados que su contraparte.Además podemos ver que a medida que incrementamos el N el modelo sigue reduciendo ambas curvas de error TrainError y TestError por lo tanto el modelo de prediccion para las elipses PARALELAS no se overfiteo como el de las DIAGONALES 
Obviamente para ambos grupos de árboles le va a afectar mucho la separación o no de las elipses. Al estar más separadas inherentemente baja el error y la complejidad media de los arboles porque es mucho más simple dividir el plano, indistintamente si sean PARALELAS O DIAGONALES las elipses

##### TP0: Diagonal y Paralelo

```python
import numpy as np
import pandas as pd
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline

#Punto A del TP, creo las clases 1 y 2, se detalla más abajo
def gaussianClassesCovMatrixDiag (d,n, C):

#Creo los centros de los cuales las clases van a "girar", d-dimensionales el (1,....,1) y el (-1,....,-1)
    clase1Center = np.ones((d,), dtype=int)
    clase2Center = np.negative(clase1Center)
#Segun el apunte de gaussianas el caso especial de las matrices diagonales y en cada elemento (i,i) de la matriz
#está la varianza( dev estandar cuadrado) en cada variable, en este caso es sobre cada coordenada. Recordemos que al ser
# una matrix de cov Diagonal, se va a tener un heatmap de una elipse centrada en los ejes en cada Clase o paralelos a ambos ejes.
#Si disminuimos C lo suficiente se puede apreciar(0.1)
    devEst = (C*sqrt(d)) ## SQRT GOOOD NO d*d
    cov = np.diag(np.full(d,devEst**2))
#Genero con distribucion Gaussiana n/2 puntos, con la matrix de Covarianza Diagonal, con centro en sus respectivos puntos
    puntosClase1A = np.random.multivariate_normal(clase1Center,cov,n//2)
    puntosClase1B = np.array(list(map (lambda pto: np.append(pto,"Clase1"),puntosClase1A)))

    puntosClase2A = np.random.multivariate_normal(clase2Center,cov,n//2)
    puntosClase2B = np.array(list(map (lambda pto: np.append(pto,"Clase2"),puntosClase2A)))
    
#Creo los DataFrame de ambas Clases y los devuelvo. Lo pude haber agrupado pero ne
    columnas = [("Coor"+ str(i+1)) for i in range(d)] #Hago una label para las columas sino queda feo
    columnas.append("Clase")
    #dataPandaC1 = pd.DataFrame(puntosClase1B, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
    #dataPandaC2 = pd.DataFrame(puntosClase2B, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
    dataPunto = pd.DataFrame(np.concatenate((puntosClase1B,puntosClase2B),axis = 0), columns = columnas)
    #Los retorno
#Conversion STR FLOAT en el dataFRAME
#print( type(TrainSets["Test"][0].loc[0,'Coor1'])) #Tengo el problema que se guardo como str los numeros WTF, todo por el map puto de arriba
    #dataPandaC1['Coor1'] = pd.to_numeric(dataPandaC1['Coor1'])
    #dataPandaC1['Coor2'] = pd.to_numeric(dataPandaC1['Coor2'])
    
    #dataPandaC2['Coor1'] = pd.to_numeric(dataPandaC2['Coor1'])
    #dataPandaC2['Coor2'] = pd.to_numeric(dataPandaC2['Coor2'])
    dataPunto['Coor1'] = pd.to_numeric(dataPunto['Coor1'])
    dataPunto['Coor2'] = pd.to_numeric(dataPunto['Coor2'])
    return dataPunto
```

```python
#Punto B del TP, creo las clases 1 y 2, se detalla más abajo
from math import sqrt

def gaussianClassesCovMatrixDiagC (d,n, C):

#Creo los centros de los cuales las clases van a "girar", d-dimensionales el (1,....,0) y el (-1,....,0)
    clase1Center = np.zeros((d,), dtype=int)
    clase1Center[0] = 1
    clase2Center = np.negative(clase1Center)
#Segun el apunte de gaussianas el caso especial de las matrices diagonales y en cada elemento (i,i) de la matriz
#está la varianza( dev estandar cuadrado) en cada variable, en este caso es sobre cada coordenada. Recordemos que al ser
# una matrix de cov Diagonal, se va a tener un heatmap de una elipse centrada en los ejes en cada Clase
    devEst = C
    cov = np.diag(np.full(d,devEst**2))## SQRT NOT GOOOD  alone SI d*d
#Genero con distribucion Gaussiana n/2 puntos, con la matrix de Covarianza Diagonal, con centro en sus respectivos puntos
    puntosClase1A = np.random.multivariate_normal(clase1Center,cov,n//2)
    puntosClase1B = np.array(list(map (lambda pto: np.append(pto,"Clase1"),puntosClase1A)))
    puntosClase2A = np.random.multivariate_normal(clase2Center,cov,n//2)
    puntosClase2B = np.array(list(map (lambda pto: np.append(pto,"Clase2"),puntosClase2A)))
#Creo los DataFrame de ambas Clases y los devuelvo. Lo pude haber agrupado pero ne
    columnas = [("Coor"+ str(i+1)) for i in range(d)] #Hago una label para las columas sino queda feo
    columnas.append("Clase")
    dataPunto = pd.DataFrame(np.concatenate((puntosClase1B,puntosClase2B),axis = 0), columns = columnas)
    #dataPandaC1 = pd.DataFrame(puntosClase1B, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
    #dataPandaC2 = pd.DataFrame(puntosClase2B, columns = columnas) #Hago el dataFrame con los puntos de la clase1 y con la label
#Los retorno
    #dataPandaC1['Coor1'] = pd.to_numeric(dataPandaC1['Coor1'])
    #dataPandaC1['Coor2'] = pd.to_numeric(dataPandaC1['Coor2'])
    
    #dataPandaC2['Coor1'] = pd.to_numeric(dataPandaC2['Coor1'])
    #dataPandaC2['Coor2'] = pd.to_numeric(dataPandaC2['Coor2'])
    dataPunto['Coor1'] = pd.to_numeric(dataPunto['Coor1'])
    dataPunto['Coor2'] = pd.to_numeric(dataPunto['Coor2'])
    return dataPunto
```

##### Resolucion

```python
n_dots = [125, 250, 500, 1000, 2000, 4000] #Cantidad de N
n_Test = 10000

#Dict para guardar los 20 Train sets para cada N, inicializamos DIAGONAL y PARALELO
TrainSetsDiagonal = {}
TrainSetsParalelo = {}
for N in n_dots:
    TrainSetsDiagonal[N] = []
    TrainSetsParalelo[N] = []
#TrainSets["Test"] = []

#Para cada N en N_dots creamos sus Xs_Train y Y_Train y los metemos en el dict RESPECTIVO
for N in n_dots:
    for i in  range(0,20):
        #DIAGONAL TRAIN SET
        #clase1Diagonal , clase2Diagonal = gaussianClassesCovMatrixDiag (2,N, 0.78)
        #print( type(clase1Diagonal.loc[0,'Coor1']))
        classDiagonal =  gaussianClassesCovMatrixDiag (2,N, 0.78)
        Xs_Train_Diag, Y_Train_Diag = classDiagonal.loc[:,["Coor1","Coor2"]] , classDiagonal.loc[:,["Clase"]]
        TrainSetsDiagonal[N].append((Xs_Train_Diag,Y_Train_Diag))
        #PARALELO TRAIN SET
        #clase1Paralelo , clase2Paralelo = gaussianClassesCovMatrixDiagC (2,N, 0.78)
        #print( type(clase1Diagonal.loc[0,'Coor1']))
        classParalelo = gaussianClassesCovMatrixDiagC (2,N, 0.78)
        Xs_Train_Paral, Y_Train_Paral = classParalelo.loc[:,["Coor1","Coor2"]] , classParalelo.loc[:,["Clase"]]
        TrainSetsParalelo[N].append((Xs_Train_Paral,Y_Train_Paral))

#Creamos el conjunto de pruebas... N = 10000 DIAGONAL
#clase1Diagonal , clase2Diagonal = gaussianClassesCovMatrixDiag (2,n_Test, 0.78)
classDiagonal = gaussianClassesCovMatrixDiag (2,n_Test, 0.78)
Xs_TestD, Y_TestD = classDiagonal.loc[:,["Coor1","Coor2"]] , classDiagonal.loc[:,["Clase"]]
TrainSetsDiagonal["Test"] = (Xs_TestD,Y_TestD)

#Creamos el conjunto de pruebas... N = 10000 PARALELO
#clase1Paralelo , clase2Paralelo = gaussianClassesCovMatrixDiagC (2,n_Test, 0.78)
classParalelo = gaussianClassesCovMatrixDiagC (2,n_Test, 0.78)
Xs_TestP, Y_TestP = classParalelo.loc[:,["Coor1","Coor2"]] , classParalelo.loc[:,["Clase"]]
TrainSetsParalelo["Test"] = (Xs_TestP,Y_TestP)

#Entrenamiento y Extraccion de datos DIAGONAL
TrainAndTestDataDiagonal = {} #(Array de ([(Tree, TreeSize, AccuracyTrain, AccuracyTest)])
#Entrenamiento y Extraccion de datos PARALELO
TrainAndTestDataParalelo = {} #(Array de ([(Tree, TreeSize, AccuracyTrain, AccuracyTest)])
for N in n_dots:
    TrainAndTestDataDiagonal[N] = []
    TrainAndTestDataParalelo[N] = []
    
for key in n_dots:
        graficarD = 0
        graficarP = 0
        for trainset in TrainSetsDiagonal[key]: #Tuple Xs_Train Y_Train   
           #Creo el Arbol para la DIAGONAL y lo ENTRENO
            clfDiagonal = DecisionTreeClassifier(criterion="entropy",min_impurity_decrease=0.005,random_state=0,min_samples_leaf=5)
            clfDiagonal.fit(trainset[0], trainset[1])
            #Predigo con el test
            prediction_Diagonal = clfDiagonal.predict(TrainSetsDiagonal["Test"][0])
            #Calculo el Tamaño del ARBOL
            TreeSizeDiagonal = clfDiagonal.tree_.node_count
            #Calculo el ERROR en el TEST y en el TRAIN
            AccuracyTestDiagonal = 1 - accuracy_score(TrainSetsDiagonal["Test"][1], prediction_Diagonal)
            AccuracyTrainDiagonal = 1 - accuracy_score(trainset[1], clfDiagonal.predict(trainset[0]))
            #Lo meto en el array
            TrainAndTestDataDiagonal[key].append((clfDiagonal,TreeSizeDiagonal,AccuracyTrainDiagonal,AccuracyTestDiagonal))
            #La primera de cada key la grafico
            if graficarD == 0 :
                #Creamos el Dataframe con las predicciones
                dfDiagonal = TrainSetsDiagonal["Test"][0].loc[:,["Coor1","Coor2"]] #copia sino me altera la referencia cuando fiteo
                dfDiagonal['Clase'] = prediction_Diagonal
                #Extraigo del dfDiagonal los puntos (X,Y) cuya prediccion fue Clase1
                Clase1PredictDiagonal = dfDiagonal.loc[dfDiagonal['Clase'] == 'Clase1', ["Coor1","Coor2"]]
                #Extraigo del dfDiagonal los puntos (X,Y) cuya prediccion fue Clase2
                Clase2PredictDiagonal = dfDiagonal.loc[dfDiagonal['Clase'] == 'Clase2', ["Coor1","Coor2"]]
                #Ploteamos
                pltC1Diagonal = Clase1PredictDiagonal.plot(kind='scatter', x='Coor1', y='Coor2', color='r', label='Clase1', title = "Test- Arbol Diagonal, N: "+ str(key) ) 
                pltC2Diagonal = Clase2PredictDiagonal.plot(kind='scatter', x='Coor1', y='Coor2', color='g', ax = pltC1Diagonal, label='Clase2') 
                graficarD = 1
        for trainset in TrainSetsParalelo[key]: #Tuple Xs_Train Y_Train   
           #Creo el Arbol para la PARALELO y lo ENTRENO
            clfParalelo = DecisionTreeClassifier(criterion="entropy",min_impurity_decrease=0.005,random_state=0,min_samples_leaf=5)
            clfParalelo.fit(trainset[0], trainset[1])
            #Predigo con el test
            prediction_Paralelo = clfParalelo.predict(TrainSetsParalelo["Test"][0])
            #Calculo el Tamaño del ARBOL
            TreeSizeParalelo = clfParalelo.tree_.node_count
            #Calculo el ERROR en el TEST y en el TRAIN
            AccuracyTestParalelo = 1 - accuracy_score(TrainSetsParalelo["Test"][1], prediction_Paralelo)
            AccuracyTrainParalelo = 1 - accuracy_score(trainset[1], clfParalelo.predict(trainset[0]))
            #Lo meto en el array
            TrainAndTestDataParalelo[key].append((clfParalelo,TreeSizeParalelo,AccuracyTrainParalelo,AccuracyTestParalelo))
            #La primera de cada key la grafico
            if graficarP == 0 :
                #Creamos el Dataframe con las predicciones
                dfParalelo = TrainSetsParalelo["Test"][0].loc[:,["Coor1","Coor2"]] #copia sino me altera la referencia cuando fiteo
                dfParalelo['Clase'] = prediction_Paralelo
                #Extraigo del dfDiagonal los puntos (X,Y) cuya prediccion fue Clase1
                Clase1PredictParalelo = dfParalelo.loc[dfParalelo['Clase'] == 'Clase1', ["Coor1","Coor2"]]
                #Extraigo del dfDiagonal los puntos (X,Y) cuya prediccion fue Clase2
                Clase2PredictParalelo = dfParalelo.loc[dfParalelo['Clase'] == 'Clase2', ["Coor1","Coor2"]]
                #Ploteamos
                pltC1Paralelo = Clase1PredictParalelo.plot(kind='scatter', x='Coor1', y='Coor2', color='r', label='Clase1', title = "Test- Arbol Paralelo, N: "+ str(key) ) 
                pltC2Paralelo = Clase2PredictParalelo.plot(kind='scatter', x='Coor1', y='Coor2', color='g', ax = pltC1Paralelo, label='Clase2') 
                graficarP = 1
            
#Graficamos el TrainERROR para la DIAGONAL
#Extraemos los datos de la DIAGONAL y PARALELO para armar los data FRAMEs y plotearlos
    #De DIAGONAL
ErrorTrainDiagonal = []
ErrorTestDiagonal = []
SizeTreeDiagonal = []
    #De PARALELO
ErrorTrainParalelo = []
ErrorTestParalelo = []
SizeTreeParalelo = []
for N in n_dots:
    ErrorTrainMeanD = []
    ErrorTestMeanD = []
    SizeTreeMeanD = []
    ErrorTrainMeanP = []
    ErrorTestMeanP = []
    SizeTreeMeanP = []
    for treetuple in TrainAndTestDataDiagonal[N]:
        ErrorTrainMeanD.append(treetuple[2]) 
        ErrorTestMeanD.append(treetuple[3]) 
        SizeTreeMeanD.append(treetuple[1])
    for treetuple in TrainAndTestDataParalelo[N]:
        ErrorTrainMeanP.append(treetuple[2]) 
        ErrorTestMeanP.append(treetuple[3]) 
        SizeTreeMeanP.append(treetuple[1])
    ErrorTrainDiagonal.append((N,np.mean(ErrorTrainMeanD))) #(ErrorTrain, N)
    ErrorTestDiagonal.append((N,np.mean(ErrorTestMeanD))) #(ErrorTest, N)
    SizeTreeDiagonal.append((N,np.mean(SizeTreeMeanD))) #(SizeTree,N)
    ErrorTrainParalelo.append((N,np.mean(ErrorTrainMeanP))) #(ErrorTrain, N)
    ErrorTestParalelo.append((N,np.mean(ErrorTestMeanP))) #(ErrorTest, N)
    SizeTreeParalelo.append((N,np.mean(SizeTreeMeanP))) #(SizeTree,N)
    
#Creo los DATAFRAME PARA PLOTEAR LOS GRAFICOS
    #Ploteo el grafico 1 DIAGONAL
columnas1 = ["N","Error Train"] #Hago una label para las columas sino queda feo
dataErrorTrainDiagonal = pd.DataFrame([[x[0],x[1]] for x in ErrorTrainDiagonal], columns = columnas1)

columnas2 = ["N","Error Test"] #Hago una label para las columas sino queda feo
dataErrorTestDiagonal = pd.DataFrame([[x[0],x[1]] for x in ErrorTestDiagonal], columns = columnas2)

pltEstadisticaDiagonal1A = dataErrorTrainDiagonal.plot(kind='line', x='N', y='Error Train', color='r', label='Error Train', title = "Error Train-Test con respecto N DIAGONAL " ) 
pltEstadisticaDiagonal1B = dataErrorTestDiagonal.plot(kind='line', x='N', y='Error Test', color='g', label='Error Test',ax = pltEstadisticaDiagonal1A ) 
    #Ploteo el grafico 1 PARALELO         
dataErrorTrainParalelo = pd.DataFrame([[x[0],x[1]] for x in ErrorTrainParalelo], columns = columnas1)
dataErrorTestParalelo = pd.DataFrame([[x[0],x[1]] for x in ErrorTestParalelo], columns = columnas2)    

pltEstadisticaParalelo1A = dataErrorTrainParalelo.plot(kind='line', x='N', y='Error Train', color='r', label='Error Train', title = "Error Train-Test con respecto N PARALELO " ) 
pltEstadisticaParalelo1B = dataErrorTestParalelo.plot(kind='line', x='N', y='Error Test', color='g', label='Error Test',ax = pltEstadisticaParalelo1A ) 

    #Ploteo el grafico 2 DIAGONAL-PARALELO MIXTOS
columnas3 = ["N","Tamaño del Tree"] #Hago una label para las columas sino queda feo
dataSizeTreeDiagonal = pd.DataFrame([[x[0],x[1]] for x in SizeTreeDiagonal], columns = columnas3)
dataSizeTreeParalelo = pd.DataFrame([[x[0],x[1]] for x in SizeTreeParalelo], columns = columnas3)
pltEstadisticaDiagonal1C = dataSizeTreeDiagonal.plot(kind='line', x='N', y='Tamaño del Tree', color='b', label='Size Tree DIAGONAL',title = "Tamaño del TREE con respecto N DIAGONAL-PARALELO " ) 
pltEstadisticaParalelo1C = dataSizeTreeParalelo.plot(kind='line', x='N', y='Tamaño del Tree', color='g', label='Size Tree PARALELO', ax = pltEstadisticaDiagonal1C  ) 
```

### EJERCICIO 3.1 CLASIFICADOR BAYES
Mi clasificador ideal tanto para las elipses DIAGONALES como para las PARALELAS... se trata de un clasificador gausiano para cada clasificador construimos el centro (el mismo que las elipses 1 por cada clase), luego construimos la matriz de covarianza segun sea el caso, luego para ambas clases con estos datos construimos la distribucion de probabilidades. Entonces dado un punto d-dimensional vamos a predecir su clase segun a qué distribución le da probabilidad mayor. Basicamente las clases se modelan como una gaussiana cada una aprovechandome que sé los centros y la matriz de covarianza saco la probabilidad para el punto y segun la que de mayor con esa me quedo. 
Esto esta bueno por 2 cosas, a más espaciadas las gaussianas de cada clase baja el error porque va a estar más clara la distribución de probabilidades y que no se intersecan o muy poco( se puede ver bajando el C). Ahora esta bueno este modelado porque a muchos pero muchos N la mayoria de los errores de clasificación se van a centrar donde las gaussianas se intersecan porque en esa area los puntos tienen cierta dualidad y se puede fallar. Y ahora esta aproximación se puede decir que es la mínima o se aproxima al mínimo error? sí, como estamos en la zona de interseccion de ambas gaussiana y siempre vamos a agarrar la que esté por encima para un mismo punto (la que da mayor prob de pertenencia a una Clase) y si tenemos muchos puntos en esta zona (demasiados inf) podemos aseverar que nos vamos a equivocar en la prediccion en la interseccion una "cantidad minima" de veces... y ese minimo tiende a ser el punto donde se intersecan las gaussianas
Espero haber sido relativamente claro

##### Resolucion

```python
from scipy.stats import multivariate_normal

#DIAGONAL BAYES CLASIFICADOR____________________________________________________
def ClasificadorBayesCovMatrixDiag (d, C, puntosDataFrama): #Punto es un [Array]
    #Centros de la gaussiana
    clase1Center = np.ones((d,), dtype=int)
    clase2Center = np.negative(clase1Center)
    #Matrix Covariance
    devEst = (C*sqrt(d))
    cov = np.diag(np.full(d,devEst**2))
    
    gaussianasParaClase1 = multivariate_normal(mean=clase1Center, cov=cov)
    gaussianasParaClase2 = multivariate_normal(mean=clase2Center, cov=cov)

    #En el map uso el .values del dataframe
    prediction = list(map(lambda punto: "Clase1" if  gaussianasParaClase1.pdf(punto) >= gaussianasParaClase2.pdf(punto) else "Clase2",puntosDataFrama.values))
    #prediction = list(map(lambda punto: "Clase1" if gaussianasParaClase1.pdf(punto) > gaussianasParaClase2.pdf(punto) else ("Clase2" if gaussianasParaClase2.pdf(punto) > gaussianasParaClase1.pdf(punto) else ("Class1" if 0.5 > 0.8 "Class2")),puntosDataFrama.values))
         

    columnas = ["Clase"] #Retorno el DataFrame
    dataPrediction = pd.DataFrame(prediction, columns = columnas)
    return dataPrediction

#PARALLEL BAYES CLASIFICADOR_____________________________________________________
def ClasificadorBayesCovMatrixDiagC (d, C, puntosDataFrama): #Punto es un [Array]
    #Centros de la gaussiana
    clase1Center = np.zeros((d,), dtype=int)
    clase1Center[0] = 1
    clase2Center = np.negative(clase1Center)
    #Matrix Covariance
    devEst = C
    cov = np.diag(np.full(d,devEst**2))
        
    gaussianasParaClase1 = multivariate_normal(mean=clase1Center, cov=cov)
    gaussianasParaClase2 = multivariate_normal(mean=clase2Center, cov=cov)
    
    #En el map uso el .values del dataframe
    prediction = list(map(lambda punto: "Clase1" if  gaussianasParaClase1.pdf(punto) >= gaussianasParaClase2.pdf(punto) else "Clase2",puntosDataFrama.values))

    columnas = ["Clase"]
    dataPrediction = pd.DataFrame(prediction, columns = columnas)
    return dataPrediction

#ErrorScore: 1- Accuracy. Funcion que calcula el error entre el dataframe original y el predictivo
def ErrorScoreCalculator(df1,df1Prediction):
    TPTN = 0
    if df1.values.shape ==  df1Prediction.values.shape and len(df1.values) == len(df1Prediction.values) :
        for i in range(0,len(df1.values)):
            if(np.array_equal(df1.values[i],df1Prediction.values[i])): #Si cada array que representa a un punto son iguales
                TPTN += 1.0
        return 1.0 - (TPTN / len(df1.values)) #Entre parens el ACCURACY
    
    else:
        return -1.0
    
#Funcion que Combina el Clasificador Ideal para Diagonales y Paralelos segun un parametro y devuelve el error
def ClasificarConBayesYObtenerError(dimensionBayes,dots_Bayes,C_Bayes, diagonalOparalelo = 0):
    listaDeCoor = [] #Lista de Coor para el DF.loc sino cualquiera sea el dimensionBayes siempre saca las dos primeras rompiendo el shape
    for d in range(0,dimensionBayes): 
        listaDeCoor.append("Coor"+str(d+1))
        
    if diagonalOparalelo == 0: #Para el DIAGONAL
        #Creamos el dataTest para la DIAGONAL
        #clase1DiagonalTest , clase2DiagonalTest = gaussianClassesCovMatrixDiag (dimensionBayes,dots_Bayes, C_Bayes)
        classDiagonalTest = gaussianClassesCovMatrixDiag (dimensionBayes,dots_Bayes, C_Bayes)
        Xs_TestD, Y_TestD = classDiagonalTest.loc[:,listaDeCoor] , classDiagonalTest.loc[:,["Clase"]]        
        #Predecimos los Xs del Test Set
        predictionDiagonalBayes = ClasificadorBayesCovMatrixDiag(dimensionBayes,C_Bayes,Xs_TestD)
        #Calculamos el error de la prediccion contra el Y_Test
        ErrorDiagonalBayes = ErrorScoreCalculator(Y_TestD,predictionDiagonalBayes)
        #Retornamos el error
        return ErrorDiagonalBayes
    else:  #Para el PARALELO
        #Creamos el dataTest para lo PARALELO
        #clase1ParaleloTest , clase2ParaleloTest = gaussianClassesCovMatrixDiagC (dimensionBayes,dots_Bayes, C_Bayes)
        classParaleloTest = gaussianClassesCovMatrixDiagC (dimensionBayes,dots_Bayes, C_Bayes)
        Xs_TestP, Y_TestP = classParaleloTest.loc[:,listaDeCoor] , classParaleloTest.loc[:,["Clase"]]
        #Predecimos los Xs del Test Set
        predictionParaleloBayes = ClasificadorBayesCovMatrixDiagC(dimensionBayes,C_Bayes,Xs_TestP)
        #Calculamos el error de la prediccion contra el Y_Test
        ErrorParaleloBayes = ErrorScoreCalculator(Y_TestP,predictionParaleloBayes)
        #Retornamos el error
        return ErrorParaleloBayes

#Por ejemplo con D=2, N = 10000, C = 0.78 en paralelas y diagonales...
print("Error Minimo Estimado Para un Conjunto Generado con DIAGONAL:")
print(ClasificarConBayesYObtenerError(5,10000,0.5,diagonalOparalelo = 0)) #Diagonal
print("Error Minimo Estimado Para un Conjunto Generado con PARALELO:")
print(ClasificarConBayesYObtenerError(5,10000,0.78,diagonalOparalelo = 1)) #Paralelo
```

### Ejercicio 3: Resistencia a Ruidos
Casi siempre el error minimo teórico va a estar por arriba del trainError ya que nosotros lo usamos para fitear nuestro modelo y por abajo generalmente del testError...Podemos ver que a más elevados los niveles de overlapping más ruido van a tener los conjuntos de entrenamiento de los modelos ya que los puntos de ambas clases van a estar mucho más superpuestos dificultando el entrenamiento. Este ruido además de notarse a nivel de entrenamiento tambien se va a expresar a nivel del testeo disparando aun mas la brecha de los valores.
Los clasificadores ideales que definimos tienen mejor resistencia a ruido que los modelos de las elipses pero aún así se puede ver que a mayor C hace que se incremente nuestra aproximación del error minimo posible. Además recordemos como explicamos en el punto 3.1 los clasificadores ideales postulados modelan las elipse como gaussianas por lo tanto si hay suficiente superposición de las clases la interseccion de las gaussianas modeladas va a ser mucho más grande quedando dificil la clasificación

```python
c_dots = [0.5, 1.0, 1.5, 2.0, 2.5] #Dimensiones
n_Test = 10000

#Dict para guardar los 20 Train sets para cada D, inicializamos DIAGONAL y PARALELO
TrainSetsDiagonal = {}
TrainSetsParalelo = {}
for C in c_dots:
    TrainSetsDiagonal[C] = []
    TrainSetsParalelo[C] = []
#TrainSets["Test"] = []

#Para cada D en D_dots creamos sus Xs_Train y Y_Train y los metemos en el dict RESPECTIVO
for C in c_dots:
    for i in  range(0,20):
        #DIAGONAL TRAIN SET
        #clase1Diagonal , clase2Diagonal = gaussianClassesCovMatrixDiag (5,250, C)
        #print( type(clase1Diagonal.loc[0,'Coor1']))
        classDiagonal = gaussianClassesCovMatrixDiag (5,250, C)
        Xs_Train_Diag, Y_Train_Diag = classDiagonal.loc[:,["Coor1","Coor2"]] , classDiagonal.loc[:,["Clase"]]
        TrainSetsDiagonal[C].append((Xs_Train_Diag,Y_Train_Diag))
        #PARALELO TRAIN SET
        #clase1Paralelo , clase2Paralelo = gaussianClassesCovMatrixDiagC (5,250, C)
        #print( type(clase1Diagonal.loc[0,'Coor1']))
        classParalelo = gaussianClassesCovMatrixDiagC (5,250, C)
        Xs_Train_Paral, Y_Train_Paral = classParalelo.loc[:,["Coor1","Coor2"]] , classParalelo.loc[:,["Clase"]]
        TrainSetsParalelo[C].append((Xs_Train_Paral,Y_Train_Paral))
    #Creamos el conjunto de pruebas... N = 10000 DIAGONAL. UNO PARA CADA C
    #clase1DiagonalTest , clase2DiagonalTest = gaussianClassesCovMatrixDiag (5,n_Test, C)
    classDiagonalTest = gaussianClassesCovMatrixDiag (5,n_Test, C)
    Xs_TestD, Y_TestD = classDiagonalTest.loc[:,["Coor1","Coor2"]] , classDiagonalTest.loc[:,["Clase"]]
    TrainSetsDiagonal["Test"+str(C)] = (Xs_TestD,Y_TestD)
    #Creamos el conjunto de pruebas... N = 10000 PARALELO. UNO PARA CADA C
    #clase1ParaleloTest , clase2ParaleloTest = gaussianClassesCovMatrixDiagC (5,n_Test, C)
    classParaleloTest = gaussianClassesCovMatrixDiagC (5,n_Test, C)
    Xs_TestP, Y_TestP = classParaleloTest.loc[:,["Coor1","Coor2"]] , classParaleloTest.loc[:,["Clase"]]
    TrainSetsParalelo["Test"+str(C)] = (Xs_TestP,Y_TestP)

#Entrenamiento y Extraccion de datos DIAGONAL
TrainAndTestDataDiagonal = {} #(Array de ([(Tree, TreeSize, AccuracyTrain, AccuracyTest)])
#Entrenamiento y Extraccion de datos PARALELO
TrainAndTestDataParalelo = {} #(Array de ([(Tree, TreeSize, AccuracyTrain, AccuracyTest)])
for C in c_dots:
    TrainAndTestDataDiagonal[C] = []
    TrainAndTestDataParalelo[C] = []
    
for key in c_dots: #No se va a trabajos con los Train
        for trainset in TrainSetsDiagonal[key]: #Tuple Xs_Train Y_Train   
           #Creo el Arbol para la DIAGONAL y lo ENTRENO
            clfDiagonal = DecisionTreeClassifier(criterion="entropy",min_impurity_decrease=0.005,random_state=0,min_samples_leaf=5)
            clfDiagonal.fit(trainset[0], trainset[1])
            #Predigo con el test
            prediction_Diagonal = clfDiagonal.predict(TrainSetsDiagonal["Test"+str(key)][0])
            #Calculo el Tamaño del ARBOL
            TreeSizeDiagonal = clfDiagonal.tree_.node_count
            #Calculo el ERROR en el TEST y en el TRAIN
            AccuracyTestDiagonal = 1 - accuracy_score(TrainSetsDiagonal["Test"+str(key)][1], prediction_Diagonal)
            AccuracyTrainDiagonal = 1 - accuracy_score(trainset[1], clfDiagonal.predict(trainset[0]))
            #Lo meto en el array
            TrainAndTestDataDiagonal[key].append((clfDiagonal,TreeSizeDiagonal,AccuracyTrainDiagonal,AccuracyTestDiagonal))
        for trainset in TrainSetsParalelo[key]: #Tuple Xs_Train Y_Train   
           #Creo el Arbol para la PARALELO y lo ENTRENO
            clfParalelo = DecisionTreeClassifier(criterion="entropy",min_impurity_decrease=0.005,random_state=0,min_samples_leaf=5)
            clfParalelo.fit(trainset[0], trainset[1])
            #Predigo con el test
            prediction_Paralelo = clfParalelo.predict(TrainSetsParalelo["Test"+str(key)][0])
            #Calculo el Tamaño del ARBOL
            TreeSizeParalelo = clfParalelo.tree_.node_count
            #Calculo el ERROR en el TEST y en el TRAIN
            AccuracyTestParalelo = 1 - accuracy_score(TrainSetsParalelo["Test"+str(key)][1], prediction_Paralelo)
            AccuracyTrainParalelo = 1 - accuracy_score(trainset[1], clfParalelo.predict(trainset[0]))
            #Lo meto en el array
            TrainAndTestDataParalelo[key].append((clfParalelo,TreeSizeParalelo,AccuracyTrainParalelo,AccuracyTestParalelo))
            
#Graficamos el TrainERROR para la DIAGONAL
#Extraemos los datos de la DIAGONAL y PARALELO para armar los data FRAMEs y plotearlos
    #De DIAGONAL
ErrorTrainDiagonal = []
ErrorTestDiagonal = []
    #De PARALELO
ErrorTrainParalelo = []
ErrorTestParalelo = []
for C in c_dots:
    ErrorTrainMeanD = []
    ErrorTestMeanD = []
    ErrorTrainMeanP = []
    ErrorTestMeanP = []
    for treetuple in TrainAndTestDataDiagonal[C]:
        ErrorTrainMeanD.append(treetuple[2]) 
        ErrorTestMeanD.append(treetuple[3]) 
    for treetuple in TrainAndTestDataParalelo[C]:
        ErrorTrainMeanP.append(treetuple[2]) 
        ErrorTestMeanP.append(treetuple[3]) 
    ErrorTrainDiagonal.append((C,np.mean(ErrorTrainMeanD))) #(ErrorTrain, N)
    ErrorTestDiagonal.append((C,np.mean(ErrorTestMeanD))) #(ErrorTest, N)
    ErrorTrainParalelo.append((C,np.mean(ErrorTrainMeanP))) #(ErrorTrain, N)
    ErrorTestParalelo.append((C,np.mean(ErrorTestMeanP))) #(ErrorTest, N)
#print("Joyita")

#OPTIMUS PRIME: Calcula la CURVA DE ERROR MINIMO, para un D y un C, aproximo el minimo error por un N muy grande
columnasErrorMin = ["C", "Error Minimo"]
dataMinErrorDiagonal = pd.DataFrame([[C,ClasificarConBayesYObtenerError(5,n_Test,C,diagonalOparalelo = 0)] for C in c_dots],columns =columnasErrorMin)
dataMinErrorParalelo = pd.DataFrame([[C,ClasificarConBayesYObtenerError(5,n_Test,C,diagonalOparalelo = 1)] for C in c_dots],columns =columnasErrorMin)


#Creo los DATAFRAME PARA PLOTEAR LOS GRAFICOS
    #Ploteo el grafico 1 DIAGONAL

columnas1 = ["C","Error Train"] #Hago una label para las columas sino queda feo
dataErrorTrainDiagonal = pd.DataFrame([[x[0],x[1]] for x in ErrorTrainDiagonal], columns = columnas1)

columnas2 = ["C","Error Test"] #Hago una label para las columas sino queda feo
dataErrorTestDiagonal = pd.DataFrame([[x[0],x[1]] for x in ErrorTestDiagonal], columns = columnas2)

pltEstadisticaDiagonal1A = dataErrorTrainDiagonal.plot(kind='line', x='C', y='Error Train', color='r', label='Error Train Diagonal', title = "Error Train-Test con respecto C-overlapping DIAGONAL" ) 
pltEstadisticaDiagonal1B = dataErrorTestDiagonal.plot(kind='line', x='C', y='Error Test', color='g', label='Error Test Diagonal',ax = pltEstadisticaDiagonal1A ) 
pltEstadisticaDiagonal1c = dataMinErrorDiagonal.plot(kind='line', x='C', y='Error Minimo', color='b', label='MIN ERROR APROX Diagonal',ax = pltEstadisticaDiagonal1B ) 

#Ploteo el grafico 1 PARALELO         
dataErrorTrainParalelo = pd.DataFrame([[x[0],x[1]] for x in ErrorTrainParalelo], columns = columnas1)
dataErrorTestParalelo = pd.DataFrame([[x[0],x[1]] for x in ErrorTestParalelo], columns = columnas2)    

pltEstadisticaParalelo1A = dataErrorTrainParalelo.plot(kind='line', x='C', y='Error Train', color='b', label='Error Train Paralelo', title = "Error Train-Test con respecto C-overlapping PARALELO") 
pltEstadisticaParalelo1B = dataErrorTestParalelo.plot(kind='line', x='C', y='Error Test', color='y', label='Error Test Paralelo',ax = pltEstadisticaParalelo1A ) 
pltEstadisticaParalelo1c = dataMinErrorParalelo.plot(kind='line', x='C', y='Error Minimo', color='r', label='MIN ERROR APROX Paralelo',ax = pltEstadisticaParalelo1B ) 
```
